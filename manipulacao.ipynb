```python
import pandas as pd
```

```python
df = pd.read_excel('data/Domingo/Gasto_Domingo.xlsx', sheet_name='GASTO_ANUAL')
```

```python
df.head()
```

```python
# acrescentar uma coluna chamada PREFEITO. E adicionar para o periodo de 2000 a 2004 - KATIA BORN. de 2005 a 2012 - CÍCERO ALMEIDA, de 2013 a 2016 - RUI PALMEIRA e de 2017 a 2020 - JHC


df['PREFEITO'] = 'KATIA BORN'
df.loc[df['ANO'] >= 2005, 'PREFEITO'] = 'CÍCERO ALMEIDA'
df.loc[df['ANO'] >= 2013, 'PREFEITO'] = 'RUI PALMEIRA'
df.loc[df['ANO'] >= 2017, 'PREFEITO'] = 'JHC'
```
